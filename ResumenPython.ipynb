Ecuacion
3x + 9y - 10z = 24
x - 6y + 4z = -4
10x - 2y + 8z = 20

```python
import numpy as np
import matplotlib.pyplot as plt
```

Crear una matriz (A) que contenga los coeficientes de las ecuaciones, en esta matriz solo estran los coeficientes de las variables de x, y, z

```python
A= np.array([
    [3, 9, -10],
    [1, -6, 4],
    [10, -2, 8]
])
```

B contendra los valores de las variables independientes

```python
B= np.array([24, -4, 20])
```

Usando la funcion linalg.solve resolvemos el sistema de ecuaciones, esta funcion se asigna a la variable "soluciones"

```python
solucion= np.linalg.solve(A, B)

print("La solucion al sistema de ecuaciones es")
print("x=", solucion[0])
print("y=", solucion[1])
print("z=", solucion[2])
```

Graficar, para eso se crea un rango de valores para x, y, z

```python
x= np.linspace(-50, 50, 100)
y= np.linspace(-50, 50, 100)
x, y = np.meshgrid(x, y)  #Crear la malla de coordenadas
```

Despejar las ecuaciones en funcion de Z para hayar los planos de la ecuacion

```python
z1= (24 - 3*x - 9*y)/-10
z2= (-4 - x + 6*y)/4
z3= (20 - 10*x + 2*y)/8
```

```python
fig = plt.figure ()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z1, alpha=0.5, rstride=100, cstride=100, label='3x+9y-10z=24')
ax.plot_surface(x, y, z2, alpha=0.5, rstride=100, cstride=100, label='x+6y+4z=-4')
ax.plot_surface(x, y, z3, alpha=0.5, rstride=100, cstride=100, label='10x-2y+8z=20')

#Graficar la solucion

ax.scatter(solucion[0], solucion[1], solucion [2], color='red', label='Solucion (x, y ,z)') #Graficar las soluciones

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#plt.legend()

plt.show()
```
